```python
import cv2
import argparse
import numpy as np
```

```python
def zoom(imgin , x_cord ,y_cord ,s):
    img=cv2.imread(imgin)
    h,w = img.shape[:2]
    print("h",h)
    print("w",w)
    print(img.shape[0])
    print(img.shape[1])
    t1 = int(x_cord - (float(1)/float(s) * float(1)/float(2)*img.shape[0]))
    t2 = int(y_cord - (float(1)/float(s)*float(1)/float(2)*img.shape[1]))
    br_x = int(t1 + (float(1)/float(s) * img.shape[0]))
    br_y = int(t2 + float(1)/float(s)*img.shape[1])
    print('t1',t1)
    print('t2',t2)
    print('br_y',br_y)
    print('br_x',br_x)
    if(t1 < 0):
        br_x = br_x - t1
    if (t2 <0):
        br_y = br_y -t2
    if (br_x > img.shape[1]):
        t1= t1 - (br_x - img.shape[1])
        br_x = img.shape[1]
        print('first')
        print(br_x)
        print(t1)
    if (br_y > img.shape[0]):
        t2 = t2 - (br_y - img.shape[0])
        br_y = img.shape[0]
    roi=img[t2:br_y, t1:br_x,:]
    wid1=(roi.shape[1])-1
    hgh1=(roi.shape[0])-1
    print("width of roi",wid1)
    print("height of roi", hgh1)
    wid_ratio = float(wid1)/float(w)
    hgh_ratio = float(hgh1)/float(h)
    print('wid_ratio',wid_ratio)
    print('hgh_ratio',hgh_ratio)
    count = 0
    new =[]
    for i in range((h)):
        for j in range((w)):
            x = int(wid_ratio*j)
            y = int(hgh_ratio*i)
            x_diff = (wid_ratio*j) - x
            y_diff = (hgh_ratio*i) - y

            # The neighbouring co-ordinates are compared
            # The 'if-else' conditions are added to check for the extreme co-ordinates
            if (x>=(w-1) or y>=(h-1)):
                A_blue = roi[y][x][0]
                A_red = roi[y][x][1]
                A_green = roi[y][x][2]
            else:
                A_blue = roi[y][x][0]
                A_red = roi[y][x][1]
                A_green = roi[y][x][2]

            if ((x+1)>=(w-1) or (y>=(h-1))):
                B_blue = roi[y][x][0]
                B_red = roi[y][x][1]
                B_blue = roi[y][x][2]
            else:
                B_blue = roi[y+1][x][0] & 0xff
                B_red = roi[y+1][x][1]
                B_green = roi[y+1][x][2]
            if (x>=(w-1) or ((y+1)>=(h-1))):
                C_blue = roi[y][x][0]
                C_red = roi[y][x][1]
                C_green = roi[y][x][2]
            else:
                C_blue = roi[y][x+1][0] & 0xff
                C_red = roi[y][x+1][1]
                C_green = roi[y][x+1][2]
            if ((x+1)>=(w-1) or (y+1)>=(h-1)):
                D_blue = roi[y][x][0] & 0xff
                D_red = roi[y][x][1]
                D_green = roi[y][x][2]
            else:
                D_blue = roi[y+1][x+1][0] & 0xff
                D_red = roi[y+1][x+1][1]
                D_green = roi[y+1][x+1][2]

            # Combining all the different channelsn into overall 3 channels
            newimg_blue = (int) ( (A_blue * (1 - x_diff) * (1 - y_diff)) + (B_blue * (x_diff) * (1 - y_diff)) + (C_blue * (y_diff)*(1 - x_diff)) + (D_blue * (x_diff*y_diff)))
            newimg_red = (int) ( (A_red * (1 - x_diff) * (1 - y_diff)) + (B_red * (x_diff) * (1 - y_diff)) + (C_red * (y_diff)*(1 - x_diff)) + (D_red * (x_diff*y_diff)))
            newimg_green = (int) ( (A_green * (1 - x_diff) * (1 - y_diff)) + (B_green * (x_diff) * (1 - y_diff)) + (C_green * (y_diff)*(1 - x_diff)) + (D_green * (x_diff*y_diff)))

            # Adding the values into the array
            newrow=count/(w)
            newcol=count%(w)
            newimg = [newimg_blue,newimg_red,newimg_green]
            if(newcol == 0):
                new.append([])
            new[newrow].append(newimg)
            count+=1
    final_img=np.uint8(new)
    return  final_img
```
